EPSILON-GREEDY

```python
#eps = 0
k = 10                                        # main
eps = 0.0
#print(r.shape)
q = np.random.normal(0,1,10)
Rew = np.zeros(shape = [1,1000])
Oa = np.zeros(shape = [1,1000]) #
for i in range(2000):
    Q = np.zeros(k)
    N = np.zeros(k)
    r = np.zeros(shape=[1,1000])       #reward
    opt = np.zeros(shape = [1,1000]) #optimal action
    
    for j in range(1000):
        p = np.random.uniform(0,1)
        if(p<eps):
            a = np.random.randint(1,10)
        else:
            a = np.argmax(Q)
            opt[0,j]  = 1            #
        
        R    =  np.random.normal(q[a],1)        #scalar value
        N[a] = N[a]+1
        Q[a] = Q[a]+(R-Q[a])/N[a]
        r[0,j] = R        
    #print(r.shape)
    #print(Rew.shape)
    #r = np.expand_dims(Re,axis = 0)
    #Rew = np.expand_dims(Rew,axis = 0)
    Rew = np.concatenate((Rew,r),axis = 0)
    Oa = np.concatenate((Oa,opt),axis = 0)
    #print(Rew.shape)
Rew0 = np.mean(Rew,axis = 0)
OptA0 = np.mean(Oa,axis = 0)
#print(Rew01.shape)

#Rew = np.mean
#Rewa = np.mean(Rew,axis =1)     
a = range(1,1001)
plt.plot(a,Rew0)
plt.plot(a,OptA0)
```

Softmax action selection method using the Gibbs distribution

```python
def act(P):                 #Function for giving index of optimal action
    global A
    a = np.random.uniform(0,1)
    #print(a)
     
    if(a<P[0]):
        A = 0
    elif(P[0]<a<P[0]+P[1]):
        A = 1
    elif(P[0]+P[1]<a<P[0]+P[1]+P[2]):
        A = 2
    elif(P[0]+P[1]+P[2]<a<P[0]+P[1]+P[2]+P[3]):
        A = 3
    elif(P[0]+P[1]+P[2]+P[3]<a<P[0]+P[1]+P[2]+P[3]+P[4]):
        A = 4
    elif(P[0]+P[1]+P[2]+P[3]+P[4]<a<P[0]+P[1]+P[2]+P[3]+P[4]+P[5]):
        A = 5
    elif(P[0]+P[1]+P[2]+P[3]+P[4]+P[5]< a < P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]):
        A = 6
    elif(P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]< a <P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]+P[7]):
        A = 7
    elif(P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]+P[7]< a <P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]+P[7]+P[8]):
        A = 8
    elif(P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]+P[7]+P[8]< a <P[0]+P[1]+P[2]+P[3]+P[4]+P[5]+P[6]+P[7]+P[8]+P[9]):
        A = 9
    return(A)

        
    
    #print(A)
        
    
    
```

```python
def softmax(x):
    "Compute softmax values for each sets of scores in x."
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
k = 10                                        # T = 30
#print(r.shape)
q = np.random.normal(0,1,10)
Rew = np.zeros(shape = [1,1000])
Oa = np.zeros(shape = [1,1000])
for i in range(2000):
    Q = np.zeros(k)
    N = np.zeros(k)
    T = 30                    #temperature constant
    r = np.zeros(shape=[1,1000])
    opt = np.zeros(shape = [1,1000])
    for j in range(1000):
        P = softmax(Q/T)
        a = act(P)
        opt[0,j]  = 1
        R   =  np.random.normal(q[a],1)        #scalar value
        N[a] = N[a]+1
        Q[a] = Q[a]+(R-Q[a])/N[a]
        r[0,j] = R        
    
    #r = np.expand_dims(Re,axis = 0)
    #Rew = np.expand_dims(Rew,axis = 0)
    Rew = np.concatenate((Rew,r),axis = 0)
    Oa = np.concatenate((Oa,opt),axis = 0)
    #print(Rew.shape)
REw1 = np.mean(Rew,axis = 0)
OptA1 = np.mean(Oa,axis = 0)

#Rew = np.mean
#Rewa = np.mean(Rew,axis =1)     
a = range(1,1001)
plt.plot(a,REw1,label = 'Avg Rewards')
plt.plot(a,OptA1, label = '%Optimal actions')
plt.legend()
plt.show()
```

UCB1

```python
k = 10                                        # main
#print(r.shape)
q = np.random.normal(0,1,10)    #ground truth value of each arm
A = np.argmax(q)
Rew = np.zeros(shape = [1,1000])            
Oa = np.zeros(shape = [1,1000])
for i in range(2000):
    Q = np.zeros(k)
    N = np.zeros(k)
    r = np.zeros(shape=[1,1000])
    opt = np.zeros(shape = [1,1000])      
    #Play each arm once
    for i in range(10):
        R = np.random.normal(q[i],1)
        N[i] = N[i]+1
        Q[i] = Q[i]+(R-Q[i])/N[i]
        
    #running for 1000 timesteps    
    for t in range(1000):
        
        UCB = (Q + np.sqrt(2*np.log(t)/N))
        #print(UCB)
        a = np.argmax(UCB)
        #print(a)
        if(a==A):
            opt[0,t]  = 1
        R    = np.random.normal(q[a],1)        #scalar value
        r[0,t] = R        
        N[a] = N[a]+1
        Q[a] = Q[a]+(R-Q[a])/N[a]
    
    Rew = np.concatenate((Rew,r),axis = 0)
    Oa = np.concatenate((Oa,opt),axis = 0)
    
Rewu = np.mean(Rew,axis = 0)
OptA1 = np.mean(Oa,axis = 0)

    
a = range(1,1001)
plt.plot(a,Rewu,label = 'Avg Rewards')
plt.plot(a,OptA1, label = '%Optimal actions')
plt.legend()
plt.show()
```
